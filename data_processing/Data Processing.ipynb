# Data Processing/Wrangling

In this notebook, we will demonstrate data processing and wrangling skills using the Automobile dataset.

## Import Libraries and Load Data

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
pd.options.display.max_colwidth = 1000
df = pd.read_csv('Automobile_data.csv')
df.head()
```

## Data Cleaning

### Replace '?' with NaN

```python
df.replace('?', np.nan, inplace=True)
df.head()
```

### Check for Missing Values

```python
df.isnull().sum()
```

### Handle Missing Values

```python
# Fill missing values for 'num-of-doors' with the mode
df['num-of-doors'].fillna(df['num-of-doors'].mode()[0], inplace=True)

# Fill missing values for 'bore' and 'stroke' with the mean
df['bore'] = df['bore'].astype(float)
df['stroke'] = df['stroke'].astype(float)
df['bore'].fillna(df['bore'].mean(), inplace=True)
df['stroke'].fillna(df['stroke'].mean(), inplace=True)

# Drop rows with missing values for 'horsepower' and 'peak-rpm'
df['horsepower'] = df['horsepower'].astype(float)
df['peak-rpm'] = df['peak-rpm'].astype(float)
df.dropna(subset=['horsepower', 'peak-rpm'], inplace=True)

#Drop symboling & normalized-losses - irrelevant
df.drop(['symboling','normalized-losses'], axis=1, inplace=True)
# Check for missing values again
df.isnull().sum()
```

```python
df.drop(df[df['price'].isnull()].index,axis=0,inplace=True)
```

## Data Transformation

### Convert 'city-mpg' and 'highway-mpg' to Liters per 100 km

```python
df['city-L/100km'] = 235 / df['city-mpg']
df['highway-L/100km'] = 235 / df['highway-mpg']
df.drop(['city-mpg', 'highway-mpg'], axis=1, inplace=True)
df.head()
```

### Normalize 'length', 'width', and 'height'

```python
df['length'] = df['length'] / df['length'].max()
df['width'] = df['width'] / df['width'].max()
df['height'] = df['height'] / df['height'].max()
df[['length', 'width', 'height']].head()

df.to_csv('Auto_Data_Proc.csv')
```
