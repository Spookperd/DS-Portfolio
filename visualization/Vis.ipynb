# Data Visualization

In this notebook, we will demonstrate data visualization skills using the Automobile dataset.

## Import Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Auto_Data_Proc.csv')
df.head()
```

## Visualizations

### Scatterplot of engine-size vs. price

```python
plt.scatter(df['engine-size'], df['price'])
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Engine Size vs. Price')
plt.show()
```

### Boxplot of price by body-style

```python
sns.boxplot(x='body-style', y='price', data=df)
plt.title('Price by Body Style')
plt.show()
```

### Histogram of price

```python
plt.hist(df['price'].dropna().astype(float), bins=10)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

## Conclusion

In this notebook, we demonstrated data visualization skills by creating scatterplots, boxplots, and histograms using the Automobile dataset.
